```
def q2(M):
    maxElement = M[0][0]
    maxI = 0
    maxJ = 0
    for i in range(len(M)):
        for j in range(len(M[0])):
            if M[i][j] > maxElement:
                maxElement = M[i][j]
                maxI = i
                maxJ = j
    print((maxI,maxJ))
q2([[0, 3, 2, 4], [2, 7, 5, 9], [6, 1, 2, 3]])
```

##Q4


```
    students = {"Jane":{"grade":90,"student_number":4012}, 
          "Pat":{"grade":81,"student_number":4052},
          "Denis":{"grade":70,"student_number":4089},
          "Max":{"grade":72,"student_number":4322},
          "Lara":{"grade":74,"student_number":4042}}
    
    gradeSum = 0   
    for student in students:
        gradeSum += students[student]["grade"]
    average = gradeSum/len(students)
       
 
    evalDict = {"less than the average.":[],
            "equal to the mean.":[],
            "above the average.":[]}
      
    
    for student in students:
        if students[student]["grade"] < average:
            evalDict["less than the average."].append(student)
        elif students[student]["grade"] == average:
            evalDict["equal to the mean."].append(student)
        else:
            evalDict["above the average."].append(student)
   
    
    grades = ("Grade of","Grades of")
    verbs = ("is","are")
    
    
    for statement in evalDict:
        studentCount = len(evalDict[statement])
        
        if studentCount == 0:
            continue
            
        selectVerb = verbs[studentCount > 1]

        selectGrades = grades[studentCount > 1]
        
        printStudents = (" and ".join(evalDict[statement])).replace(' and ', ', ' , studentCount-2)
        
        print(selectGrades, printStudents, selectVerb, statement )
```

```
def q5(left=200):
    sales = 0
    while(left>0):
        print(f'{left} tickets remaining')
        print('How many do you want to buy?')
        print(f'Minimum 1, maximum {min(4,left)}')
        buy = int(input())
        sales += 1
        left -= buy
        
    print(f'{sales} sales transactions')

q5()
```
